## 1.3 Calidad de los datos

```python
# Funci√≥n para identificar los duplicados

def duplicados(df, nombre,clave=None):
   print(f"***** Duplicados en {nombre} *****")
   print("_" * 50)
   print()
   dup_rows = df.duplicated(keep=False).sum()
   print("Filas duplicadas (exactas):", dup_rows)
   if clave and clave in df.columns:
        dup_key = df[clave].duplicated(keep=False).sum()
        print(f"Duplicados por clave '{clave}':", dup_key)
        if dup_key > 0:
             duplicados_ordenados = (
                df[df[clave].duplicated(keep=False)][clave]
                .value_counts()                # Cuenta cu√°ntas veces aparece cada valor
                .sort_values(ascending=False)   # Ordena de mayor a menor (m√°s duplicados arriba)
              )
             print("\nüîÅ Top valores duplicados m√°s frecuentes:")
             # Mostrar solo los primeros 10 (los m√°s repetidos)
             display(duplicados_ordenados.head(10))
        else:
            print(f"No se encontraron duplicados en la clave '{clave}'.")
   else:
        # Si la clave no fue pasada o no existe en el DataFrame
        if clave:
            print(f"La clave '{clave}' no existe en el DataFrame.")
        else:
            print("No se indic√≥ una clave para analizar duplicados por columna.")
```

```python
# Normalizo fechas

def fecha_norm (df):
  flag=True
  for col in df.columns:
        if "fecha" in col.lower():  # detecta columnas con la palabra "fecha"
            df[col] = pd.to_datetime(df[col], errors="coerce", dayfirst=True)
            print(f"Columna '{col}' normalizada a '{df[col].dtypes}'")
            flag=False
            print("-" * 50)

  if flag:
      print("No se encontraron columnas con el patr√≥n 'fecha' en el DataFrame.")
```

```python
# Funci√≥n para normalizar texto

def normalizar_texto(df):
    for col in df.select_dtypes(include="object").columns:
        df[col] = (                                 # Usar () cuando va en distintas lineas
            df[col]
            .astype(str)                              # Convierte cualquier tipo a string
            .str.strip()                               # Elimina espacios al inicio y final
            .str.replace(r"[\u200b\t\r\n]", "", regex=True) # Borra tabs, saltos, etc
            .str.replace(" +", " ", regex=True)             # Reemplaza espacios consecutivos x 1
            .str.title()                               # Convierte a T√≠tulo
        )

        #df[col] = df[col].astype(str).str.strip().str.replace(r"[\u200b\t\r\n]", "", regex=True).str.replace(" +", " ", regex=True).str.title()
    return df
```

## 2.2 Transformaci√≥n de datos

```python
# Creo una tabla de ventas que muestre solo los productos con alto rendimiento.

# Calculo el ingreso y creo un nuevo df con la columna "ingreso"
df_ventas_ingresos = df_ventas_clean.assign(ingreso = df_ventas_clean["precio"] * df_ventas_clean["cantidad"])

# Agrupo filas por producto del df_ventas_ingresos y tengo un nuevo df: resumen_prod
resumen_prod = (
    df_ventas_ingresos.groupby(by="producto",dropna=False,as_index=False,observed=False)
    .agg(
        ingreso_total=('ingreso', 'sum'),
        unidades=('cantidad', 'sum'),
        precio_promedio=('precio', 'mean'),
        registros=('ingreso', 'size'))
)

# Redondeo el 'precio_promedio'
resumen_prod['precio_promedio'] = resumen_prod['precio_promedio'].round(2)

#obtengo percentil 80 (para luego quedarme con top 20% de m√°s ingresos)
p80_ingreso=resumen_prod["ingreso_total"].quantile(q=0.8,interpolation="linear")
#print("Percentil 80: ",p80_ingreso.round(2))

# Filtro los productos de alto rendimiento y ordeno de mayor a menor
prod_alto_rendimiento = (
    resumen_prod.query("ingreso_total >= @p80_ingreso",engine="python")
    .sort_values(by=["ingreso_total","unidades"],ascending=[False,False],ignore_index=True)
)

# Muestro resultados
print()
print("**  Columna: Producto  **")
print(f"Umbral (percentil 80) de ingreso_total: {float(p80_ingreso):,.2f}")
print("‚úÖ Productos de ALTO RENDIMIENTO (top 20% por ingreso):")
display(prod_alto_rendimiento.head(20))
```

## 2.3 Agregaci√≥n

```python
# Resumo las ventas por categor√≠a de producto y analizo los ingresos generados.

# Calculo el ingreso y creo un nuevo df con la columna "ingreso"
df_ventas_cat = df_ventas_clean.assign(ingreso = df_ventas_clean["precio"] * df_ventas_clean["cantidad"])

# Agrupo filas y obtengo m√©tricas
resumen_cat = (
    df_ventas_cat.groupby(by="categoria",dropna=False,as_index=False)
    .agg(
        ingreso_total=('ingreso', 'sum'),
        unidades=('cantidad', 'sum'),
        precio_promedio=('precio', 'mean'),
        registros=('ingreso', 'size')
        )
    .sort_values(
        by='ingreso_total',
        ascending=False,
        na_position='last',
        ignore_index=False
    )
)

# Redondeo el 'precio_promedio'
resumen_cat['precio_promedio'] = resumen_cat['precio_promedio'].round(2)


# Muestro resultados
print()
print("**  Columna: Categor√≠a  **")
print("Resumen por categor√≠a (ordenado por ingreso_total):")
display(resumen_cat.head(20))
```

## 2.4 Integraci√≥n de datos:
Combinaci√≥n de los sets de datos de ventas y marketing para ver tendencias

```python
# Creo el df de ventas_x_producto desde ventas_clean agrupado por producto y
# agrego la cantidad por el precio
df_ventas_x_producto = df_ventas_clean.groupby('producto').agg(
    precio_por_cantidad=('precio', lambda x: (x * df_ventas_clean.loc[x.index,'cantidad']).sum())
)

# Ordeno por producto de forma ascendente
df_ventas_x_producto = df_ventas_x_producto.sort_values(by='producto', ascending=True)

# muestro resultado
display(df_ventas_x_producto)


```

```python
# Gr√°fico de l√≠neas utilizando Matplotlib
plt.figure(figsize=(10, 6))
plt.plot(df_ventas_x_producto.index, df_ventas_x_producto['precio_por_cantidad'], marker='s', linestyle='solid',
color='orange', linewidth=2, markersize=6)
plt.title('Total vendido por Producto', fontsize= 14)
plt.xlabel('Producto', fontsize=12)
plt.ylabel('Precio', fontsize=12)
plt.grid(True)
plt.xticks(rotation=80)
plt.show()
```

```python
# Agrupo el df_marketing_clean por producto y sumo el costo de las campa√±as de ese producto

df_marketing_x_producto = df_marketing_clean.groupby('producto')['costo'].sum().reset_index()
df_marketing_x_producto = df_marketing_x_producto[['producto', 'costo']]
df_marketing_x_producto = df_marketing_x_producto.sort_values(by='producto', ascending=True)
display(df_marketing_x_producto)
```

```python
# Gr√°fico de barras utilizando Matplotlib
plt.figure(figsize=(10, 6))
plt.bar(df_marketing_x_producto['producto'], df_marketing_x_producto['costo'], color='darkgreen', width=0.4)
plt.title('Total costo campa√±as por Producto', fontsize= 14)
plt.xlabel('Producto', fontsize=12)
plt.ylabel('Costo', fontsize=12)
#plt.grid(True)
plt.xticks(rotation=80)
plt.show()
```

```python
# Hago un merge del df de ventas por producto con el df de marketing por producto
df_merged = pd.merge(df_ventas_x_producto, df_marketing_x_producto, on='producto')
display(df_merged.sort_values(by='costo', ascending=False))
```

```python
# Obtengo la relaci√≥n costo/ventas de cada producto
df_merged['%_costo_ingreso'] = (df_merged['costo'] / df_merged['precio_por_cantidad']) * 100
display(df_merged.sort_values(by='%_costo_ingreso', ascending=False))
```

```python
# Estad√≠stica descriptiva (global) sobre columnas relevantes
def resumen_estadistico(serie, nombre):
    print(f"\n--- {nombre} ---")
    print(f"Count: {serie.count()}")
    print(f"Media: {serie.mean():.2f}")
    print(f"Mediana: {serie.median():.2f}")
    try:
        moda = serie.mode().iloc[0]
    except:
        moda = np.nan
    print(f"Moda: {moda}")
    print(f"Std: {serie.std():.2f}")
    print(f"Varianza: {serie.var():.2f}")
    print(f"Rango: {serie.max() - serie.min():.2f}")
    print(f"IQR: {serie.quantile(0.75) - serie.quantile(0.25):.2f}")
    percentiles = serie.quantile([0.25, 0.5, 0.75]).values
    print(f"Percentiles (25,50,75): [{percentiles[0]:.2f}, {percentiles[1]:.2f}, {percentiles[2]:.2f}]")


print("\n## Estad√≠stica descriptiva - INGRESO")
resumen_estadistico(df_ventas_ingresos['ingreso'], 'Ingreso')

print("\n## Estad√≠stica descriptiva - PRECIO")
resumen_estadistico(df_ventas_ingresos['precio'], 'Precio')

print("\n## Estad√≠stica descriptiva - CANTIDAD")
resumen_estadistico(df_ventas_ingresos['cantidad'], 'Cantidad')
```

### C√°lculo del rango intercuart√≠lico (IQR) y outliers de productos

```python
lim_inf = q1 - 1.5 * iqr
lim_sup = q3 + 1.5 * iqr

outliers = resumen_prod[(resumen_prod['ingreso_total'] < lim_inf)
            | (resumen_prod['ingreso_total'] > lim_sup)]

print(f"L√≠mite inferior: {lim_inf:,.2f}")
print(f"L√≠mite superior: {lim_sup:,.2f}")
print("\nOutliers detectados:")
print(outliers.sort_values(by='ingreso_total', ascending=False))
```

## 3.3 Correlaci√≥n entre precio y unidades vendidas


```python
x = resumen_prod['precio_promedio']
y = resumen_prod['unidades']

m, b = np.polyfit(x, y, 1)

plt.figure(figsize=(7, 5))
plt.scatter(x, y)

plt.plot(x, m*x + b)

plt.title('Precio promedio vs Unidades (por producto)')
plt.xlabel('Precio promedio')
plt.ylabel('Unidades vendidas')
plt.tight_layout()
plt.show()
```

```python
# Calculamos el coeficiente de correlaci√≥n de Pearson con pandas (m√©todo corr)
corr_val = resumen_prod[['precio_promedio','unidades']].corr().loc['precio_promedio','unidades']

print(f"Coeficiente de correlaci√≥n (Pearson): {corr_val:.3f}")
# Interpretaci√≥n simple:
if corr_val > 0.8:
    print("Interpretaci√≥n: relaci√≥n positiva clara (a mayor precio_promedio, suelen venderse m√°s unidades).")
elif corr_val > 0.2:
    print("Interpretaci√≥n: relaci√≥n positiva d√©bil a moderada.")
elif corr_val < -0.8:
    print("Interpretaci√≥n: relaci√≥n negativa clara (a mayor precio_promedio, suelen venderse menos unidades).")
elif corr_val < -0.3:
    print("Interpretaci√≥n: relaci√≥n negativa d√©bil a moderada.")
else:
    print("Interpretaci√≥n: relaci√≥n muy d√©bil o nula.")
```

##### An√°lisis de la correlaci√≥n entre precio y unidades vendidas
Hay una tendencia general muy d√©bilmente positiva. La l√≠nea de regresi√≥n se inclina ligeramente hacia arriba. Esto sugiere que, en promedio, a medida que el precio promedio de un producto aumenta, las unidades vendidas tambi√©n tienden a aumentar ligeramente, aunque la relaci√≥n es muy d√©bil.
La mayor√≠a de los puntos de datos est√°n dispersos lejos de la l√≠nea de regresi√≥n. Esto indica una correlaci√≥n baja entre las dos variables. Hay muchos otros factores que probablemente influyen en las ventas, no s√≥lo el precio promedio. Las unidades vendidas var√≠an considerablemente para rangos de precios similares.
```python
# An√°lisis de campa√±as: Cruce Ventas <-> Marketing

ventas=df_ventas_ingresos.copy()
marketing=df_marketing_clean.copy()
merged = ventas.merge(marketing, on='producto', how='left', suffixes=('_venta','_camp'))

# Creo columna 'en_campa√±a' donde la fecha_venta est√© entre fecha_inicio y fecha_fin
cond = (merged['fecha_venta'] >= merged['fecha_inicio']) & (merged['fecha_venta'] <= merged['fecha_fin'])
merged['en_campa√±a'] = cond.fillna(False)  #si falta alguna fecha asume que es False

# Ventas que cayeron dentro de alguna campa√±a: (recordar... el [] filtra! las que tienen True)
ventas_en_campa√±a = merged[merged['en_campa√±a']].copy()

# Ventas que no est√°n asociadas a ninguna campa√±a (fecha fuera de periodos o no hay campa√±a para ese producto)

print("\nVentas totales:", len(ventas))
print("Ventas asociadas a campa√±a (coincidencias por producto+fecha):", len(ventas_en_campa√±a))
```

```python
# M√©tricas: % de ventas que ocurrieron durante campa√±as (por n√∫mero de transacciones y por ingreso)
pct_ventas_count = len(ventas_en_campa√±a) / len(ventas) * 100
pct_ventas_ingreso = ventas_en_campa√±a['ingreso'].sum() / ventas['ingreso'].sum() * 100
print(f"\nPorcentaje de transacciones en campa√±a: {pct_ventas_count:.2f}%")
print(f"Porcentaje de ingreso en campa√±a: {pct_ventas_ingreso:.2f}%")
```

```python
# Rendimiento por campa√±a (ventas generadas y ROI)

ventas_por_camp = ventas_en_campa√±a.groupby(['id_campanha','producto','canal'])['ingreso'].sum().reset_index().rename(columns={'ingreso':'ventas_generadas'})

#print("ventas_por_camp")
#print(ventas_por_camp.head(5))

# Uno costo de campa√±a
camp_info = marketing[['id_campanha','costo','fecha_inicio','fecha_fin','canal','producto']].drop_duplicates(subset=['id_campanha'])
camp_perf = ventas_por_camp.merge(camp_info, on=['id_campanha','producto','canal'], how='right')

#print("\ncamp_perf")
#print(camp_perf.head(5))

# Si una campa√±a no tuvo ventas en el periodo, ventas_generadas ser√° NaN -> 0
camp_perf['ventas_generadas'] = camp_perf['ventas_generadas'].fillna(0)

# Saco ROI simple: (ventas_generadas - costo) / costo
camp_perf['roi'] = (camp_perf['ventas_generadas'] - camp_perf['costo']) / camp_perf['costo']

# Ranking de ROI
camp_perf_sorted = camp_perf.sort_values(by='roi', ascending=False)
display(camp_perf_sorted[['id_campanha','producto','canal','costo','ventas_generadas','roi']].head(15))
print("\n")

# Gr√°fico de ventas generadas por canal (solo ventas en campa√±as)
ventas_canal = ventas_en_campa√±a.groupby('canal')['ingreso'].sum().sort_values(ascending=False)
fig=px.bar(ventas_canal,y=ventas_canal.index, x=ventas_canal.values)

fig.update_layout(
    title="Ventas generadas por canal (solo datos dentro de campa√±as)",
    xaxis_title="Ingreso generado",
    yaxis_title="Canal",
    width=800,
    height=400
)
fig.show()
```
